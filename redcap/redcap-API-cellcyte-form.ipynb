## Cellcyte Image Processing Form - REDCap API

```python
import sys
import os
import re
import requests
from io import StringIO
import pandas as pd
import numpy as np
from datetime import datetime
from config import config
```

### Get Record
Export a specific experiment record to a dataframe

```python
# The cellcyte dataset id you want to export
DATASET_ID = 'ABCDEF'

api_url = config['api_url']
api_token = config['api_token']

get_record = {
    'token': api_token,
    'content': 'record',
    'format': 'csv',
    'type': 'flat',
    'filterLogic': f"[dataset_id]='{DATASET_ID}'"
}

record_response = requests.post(api_url, data=get_record)
print('get record HTTP Status: ' + str(record_response.status_code))

exp_data = StringIO(record_response.text)
exp_df = pd.read_csv(exp_data, sep=',', dtype='str')
exp_df
```

### Export uploaded plate map file

```python
get_file = {
    'token': api_token,
    'content': 'file',
    'action': 'export',
    'record': exp_df['record_id'].values[0],
    'field': 'platemap_file_upload',
    'returnFormat': 'csv'
}

file_response = requests.post(api_url,data=get_file)
print('get file HTTP Status: ' + str(file_response.status_code))

file_data = StringIO(file_response.text)
pmap = pd.read_csv(file_data, sep='\t')
pmap
```
